# Visualizacion de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de dispersion (scatterplot)

x= np.arange (2,20)
y=x**2
y2=x**3

plt.scatter(x, y, color="green", marker="*", label="X^2")
plt.scatter(x, y2, color="pink", marker="s", label="X^3")
plt.plot(x, y2, color="red", marker="s", label="line")
plt.title("Funciones polinomicas", fontsize=20, color="red") #diagrama lineas
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

plt.text(5, 4000, "texto", color="blue")
```

# histogramas

```python
# representa la distribucion de variables numericas unidimensionales
mean=5
std=3
n=1000
x=np.random.normal(mean, std, n)

plt.hist(x, bins=20);

plt.title("Histograma de una variable aleatoria", fontsize=20, color="red") #diagrama lineas
plt.xlabel("x")
plt.ylabel("frecuencia")
plt.grid()
plt.legend()

plt.text(6, 140, f"mean: {mean} +/- {std}", color="blue")
```

```python
plt.boxplot(x, vert=False);
```

```python
# subplots

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15, 5), sharex=True)

ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```
